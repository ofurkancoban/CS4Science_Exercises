3) Transitions:

Give 3 words over the FSM's alphabet, which have a length > 2 but are not in the FSM's language.

* aaa → Invalid: There is no transition for a from state B after the first a. The state B has no outgoing transition for a.

* abdd → Invalid: This word ends in state F instead of the accepting state B. The FSM needs to end in state B to be accepted.
