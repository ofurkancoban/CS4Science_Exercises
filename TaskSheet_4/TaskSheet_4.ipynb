## Task 5: Disjunctive Normal Form (4.5 points)

The Boolean expression **F** is defined by the following truth table. To model **F** by a Disjunctive Normal Form (DNF), add the minterm in the correct lines. Only fill in the lines that need to be modeled by a minterm.

| Line | A | B | C | F | Terms             |
|------|---|---|---|---|-------------------|
| 1    | 0 | 0 | 0 | 1 | ¬A ∧ ¬B ∧ ¬C       |
| 2    | 0 | 0 | 1 | 0 |                   |N
| 3    | 0 | 1 | 0 | 0 |                   |
| 4    | 0 | 1 | 1 | 1 | ¬A ∧ B ∧ C         |
| 5    | 1 | 0 | 0 | 0 |                   |
| 6    | 1 | 0 | 1 | 0 |                   |
| 7    | 1 | 1 | 0 | 0 |                   |
| 8    | 1 | 1 | 1 | 1 | A ∧ B ∧ C           |

### Connecting the Minterms

The minterms corresponding to the lines where **F = 1** are connected using the **OR** operator. 

**Disjunctive Normal Form (DNF) for F:**

\[ F = (\neg A \land \neg B \land \neg C) \lor (\neg A \land B \land C) \lor (A \land B \land C) \]

### Explanation

- **Line 1:** When A, B, and C are all **0** (False), **F** is **1** (True). The corresponding minterm is **¬A ∧ ¬B ∧ ¬C**.
- **Line 4:** When A is **0**, B and C are **1**, **F** is **1**. The corresponding minterm is **¬A ∧ B ∧ C**.
- **Line 8:** When A, B, and C are all **1**, **F** is **1**. The corresponding minterm is **A ∧ B ∧ C**.

By combining these minterms with the **OR** operator, we obtain the DNF for **F**.
